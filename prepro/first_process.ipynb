```python
import pandas as pd
import numpy as np
```

```python
def processing(df):
  df['reviewContent'] = df['reviewContent'].replace('', np.nan)
  df = df.dropna(subset=['reviewContent'])
  return df
```

```python
from pathlib import Path

df = pd.read_csv('all_reviews_combined.csv')
df = processing(df)

output_filename = f"processed_reviews.csv"
df.to_csv(output_filename, index=False)

print(f"Processed file saved as: {output_filename}")
```
